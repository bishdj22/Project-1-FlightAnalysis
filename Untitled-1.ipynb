```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
airport_csv = ("Data/final_dataset.csv")
airport_data_df = pd.read_csv(airport_csv)
airport_data_df.head()
```

```python
#airport_data_df['airport_name'].value_counts()
airline_groups = airport_data_df.groupby(['carrier_name'])['total_number_flights']
airline_delays = airline_groups.sum()

airline_df = airline_delays.reset_index()
airline_df
```
